```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("dataset/diabates.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
x = df.drop(columns = ['Outcome'])
y = df['Outcome']
```

```python
x_train, x_test,y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.fit_transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
pred = model.predict(x_test)
```

```python
print(f"Accuracy Score: {accuracy_score(y_test, pred)}")
```

```python
print(f"Classification Report: {classification_report(y_test, pred)}")
```

```python
new_patient = [[140, 28.5, 45]]
new_patient_scaled = scalar.transform(new_patient)

new_pred = model.predict(new_patient_scaled)
if new_pred[0] == 1:
    print("Diabates")
else:
    print("Non Diabates")
```
