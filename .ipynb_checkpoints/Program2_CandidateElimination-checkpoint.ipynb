```python
import pandas as pd
```

```python
data = [
    ["Sunny", "Hot", "High", "Weak", "No", "Yes"],
    ["Sunny", "Hot", "High", "Strong", "No", "Yes"],
    ["Rainy", "Mild", "High", "Weak", "No", "No"],
    ["Sunny", "Mild", "High", "Weak", "Yes", "Yes"]
]

columns = ["Sky", "AirTemp", "Humidity", "Wind", "Water", "EnjoySport"]
```

```python
df = pd.DataFrame(data, columns=columns)
df.head()
```

```python
concepts = df.iloc[:,:-1].values
target = df.iloc[:,-1].values
```

```python
s = concepts[0].copy()
g = ["?"] * len(s)
```

```python
for h, t in zip(concepts, target):
    if t == "yes":
        s= [s[i] if s[i] == h[i] else "?" for i in range(len(s))]
    else:
        g = [s[i] if s[i] != h[i] else "?" for i in range(len(s))]
```

```python
print(f"Specific Hypothesis: {s}")
print(f"General Hypothesis: {g}")
```
