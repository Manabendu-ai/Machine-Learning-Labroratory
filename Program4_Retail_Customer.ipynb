```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('dataset/retail.csv')
df.head()
```

```python
df.info()
```

```python
def prepare_raw(df):
    data = df.copy()

    le = LabelEncoder()
    for i in ['Gender','City', 'Purchase']:
        data[i] = le.fit_transform(data[i])

    x = data.drop(columns = ["CustomerId", "Purchase"])
    y = data["Purchase"]

    return train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
def prepare_clean(df):
    data = df.copy()

    data.drop_duplicates(inplace=True)

    data.fillna(data.mode().iloc[0], inplace=True)
    
    le = LabelEncoder()
    for i in ['Gender','City', 'Purchase']:
        data[i] = le.fit_transform(data[i])

    x = data.drop(columns = ["CustomerId", "Purchase"])
    y = data["Purchase"]

    return train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
def get_accuracy(x_train, x_test, y_train, y_test):
    model = DecisionTreeClassifier(random_state=42)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    return accuracy_score(y_test, pred) * 100
```

```python
x_train, x_test, y_train, y_test = prepare_raw(df)
raw_acc = get_accuracy(x_train, x_test, y_train, y_test)

x_train, x_test, y_train, y_test = prepare_clean(df)
clean_acc = get_accuracy(x_train, x_test, y_train, y_test)
```

```python
print(f"Accuracy Before Preprocessing : {raw_acc:.2f}%")
print(f"Accuracy After Preprocessing  : {clean_acc:.2f}%")
```
